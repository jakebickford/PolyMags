```python
#function to more efficiently implement the selection and processing of frequency based on keyword 
def getFreq(list, keyword):#go through frequenc terms list
    for tup in list: #for each tuple in list
        if tup[0] == keyword: #if first item in tuple matches keyword (i.e. )
            return tup[1] # return the second item in tuple (i.e. the number of occurences)
    # if it's gone through the whole list and not found it return zero
    return 0
```

```python
def addColumn (keyword):
  df[keyword] = df['Frequent_tokens'].apply(getFreq, args=[keyword]) #apply above to add new column to df with frequency of the selected keyword      
```

```python
#13/08/21 returning here to do another frequency analysis of the whole corpus to add to public notebook

import pandas as pd
import pickle
import nltk
from nltk.probability import FreqDist
df = pd.read_pickle('corpus_10.pkl')#reading most recent df - note that this requires the other notebooks to have been run to generate this version of pickled DF (i.e. corpus 10 is generated in 5.Lemmatization.ipynb)

```

```python
df
```

```python
#extract words from tokenized_words
words = []
for word in (df['tokenized_words']):
    words.append(word)

#but the above gives us a list of lists, so we need to extract the words from that
allWords = []
for wordList in words:
    allWords += wordList
```

```python

#to do the frequency analysis    
fdist = FreqDist(allWords)
```

```python
print(fdist.most_common(100))
```

```python
save_FreqDist = open("FinalFreqDist.pkl","wb") #save frequency analysis as a pickle so it can be imported to public notebook.
pickle.dump(fdist, save_FreqDist)
save_FreqDist.close()
```
