```python
#search for a keyword, and generate list of frequencies across issues
listoffreqs = []#create empty list to hold frequencies
keyword='war'#enter keyword to search for here
found = False #set boolean as false

for dflevel in df['Frequent_tokens']: #loop through each row
    found = False
    for lstlevel in dflevel:#within each row loop through list of tuples
        if lstlevel[0] == keyword:
            found = True
            listoffreqs.append(lstlevel[1])#if keyword found append the frequency number of keyword to list of frequencies
    if found == False:
        listoffreqs.append(0)#if keyword not found append 0 to the list of frequencies

df['freq'] = listoffreqs#add list of frequencies as new column to df
```

```python
#function to more efficiently implement the selection and processing of frequency based on keyword 
def getFreq(list, keyword):
    for tup in list:
        if tup[0] == keyword:
            return tup[1]
    # if it's gone through the whole list and not found it return zero
    return 0
```

```python
df[keyword] = df['Frequent_tokens'].apply(getFreq, args=[keyword])
```

```python
#13/08/21 returning here to do another frequency analysis of the whole corpus to add to public notebook

import pandas as pd
import pickle
import nltk
from nltk.probability import FreqDist
df = pd.read_pickle('corpus_10.pkl')#reading most recent df

```

```python
df
```

```python
#extract words from tokenized_words
words = []
for word in (df['tokenized_words']):
    words.append(word)

#but the above gives us a list of lists, so we need to extract the words from that
#see https://stackoverflow.com/questions/38666973/pandas-nltk-tokenizing-unhashable-type-list
allWords = []
for wordList in words:
    allWords += wordList
```

```python

#to do the frequency analysis    
fdist = FreqDist(allWords)
```

```python
print(fdist.most_common(100))
```

```python
save_FreqDist = open("FinalFreqDist.pkl","wb")
pickle.dump(fdist, save_FreqDist)
save_FreqDist.close()
```
