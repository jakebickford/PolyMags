```python
import pandas as pd
import pickle
from sklearn.feature_extraction import text 
from sklearn.feature_extraction.text import CountVectorizer

#workflow adapted from Alice Zhao (2020), https://github.com/adashofdata/nlp-in-python-tutorial/blob/master/4-Topic-Modeling.ipynb
```

```python
#read the pickled df
df = pd.read_pickle('corpus_8.pkl') #dataframe with lemmatized tokens
```

```python
df
```

```python
#having done some initial topic modelling I have identified some additional words I would like to remove as stop words:
#thee
#tue
#mon
```

```python
spelling = ['thee', 'tue', 'mon']
```

```python
df['Lemmatized_tokens'] = df['Lemmatized_tokens'].apply(lambda x: [item for item in x if item not in spelling]) #remove additional stop words from lemmatized tokens
```

```python
df['tokenized_words'] = df['tokenized_words'].apply(lambda x: [item for item in x if item not in spelling]) #doing this for the tokenized words too
```

```python
df
```

```python
#create document term matrix - first we need the lemmatized tokens as a string
```

```python
def ListToString(text): #function for converting list to string
    words_as_string = ''
    for word in text:
        str(word)
        words_as_string = words_as_string + ' ' + word
    return words_as_string
```

```python
df['Lemmatized_tokens_as_string'] = df['Lemmatized_tokens'].apply(ListToString) #call function on lemmatized tokens
```

```python
#create document-term matrix using sklearn CountVectorizer

cv = CountVectorizer(min_df=10, max_df=0.5) # create count vectorizer object. min/max def: specifies max and min number of occurances e.g. min_df=10 would only consider words which appear in at least 10 docs and max_df=0.5,would only consider words which appear in at most 50% of the docs - see https://nbviewer.jupyter.org/github/adelr/trove-refugee/blob/master/Analyses.ipynb
data_cv = cv.fit_transform(df.Lemmatized_tokens_as_string) #use count vectorizer on lemmatized tokens
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())#create document term matrix
data_dtm.index = df.index
data_dtm
```

```python
#import topic modelling modules
from gensim import matutils, models
import scipy.sparse
```

```python
#put tdm into gensim corpus
sparse_counts = scipy.sparse.csr_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
```

```python
# create dictionary of location of termsin the term-document matrix
cv = pickle.load(open("cv.pkl", "rb"))
id2word = dict((v, k) for k, v in cv.vocabulary_.items())
```

```python
from gensim import corpora

word2id = dict((k, v) for k, v in cv.vocabulary_.items())
d = corpora.Dictionary()
d.id2token = id2word
d.token2id = word2id
#added additional parameters to dictionary as pyLDAvis was not working without them, see https://stackoverflow.com/questions/59024220/error-in-computing-the-coherence-score-attributeerror-dict-object-has-no-at?rq=1
```

```python
lda = models.LdaModel(corpus=corpus, id2word=d, num_topics=20, passes=10) #lda with 20 topics, 10 passes
lda.print_topics()
```
